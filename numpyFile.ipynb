```python
import numpy as np
```

```python
# Arithematic operations: + - * /
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b) # [5 7 9]
print(a * b) # [4 10 18]
print(a / b) # [0.25 0.4 0.5]
print(a - b) # [-3 -3 -3]
```

```python
# Basic Mathematical functions
tri = np.array([0, np.pi/2, np.pi,1])
print(np.sin(tri)) # [0. 1. 0. 0.84]
print(np.cos(tri)) # [1. 0. -1. 0.54]
print(np.exp(tri)) # [1.4.81 20.09 2.72]
print(np.sqrt(tri))
```

```python
# sorting arrays
val = [("popo",2008,7.2),("koko",2007,8.9),("Dodo",2009,6.5)]
dtype = [("Name","S10"),("Year",int),("GPA",float)]

arr = np.array(val, dtype=dtype)

print(np.sort(arr, order="Name"))
print(np.sort(arr, order="Year"))
print(np.sort(arr, order="GPA"))
print(np.sort(arr, order=["Year", "GPA"]))
print(arr)
```
